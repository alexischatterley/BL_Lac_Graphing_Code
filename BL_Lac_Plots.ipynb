```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
R_data = pd.read_csv('Reduced-BLLac-Data-R.csv')
f = plt.figure()
f.set_figwidth(16)
f.set_figheight(9)
plt.errorbar(R_data.mjd, R_data.mag, R_data.mag_err, marker='o', ms=4, linestyle = "none")
plt.gca().invert_yaxis()
plt.title("R-band", fontsize=20)
plt.xlabel("Modified Julian Date", fontsize=20)
plt.ylabel("Magnitude", fontsize=20)
f.savefig("R_data.jpeg")
plt.show()
```

```python
B_data = pd.read_csv('Reduced-BLLac-Data-B.csv')
g = plt.figure()
g.set_figwidth(16)
g.set_figheight(9)
plt.errorbar(B_data.mjd, B_data.mag, B_data.mag_err, marker='o', ms=4, linestyle="none")
plt.gca().invert_yaxis()
plt.title("B-band", fontsize=30)
plt.xlabel("Modified Julian Date", fontsize=20)
plt.ylabel("Magnitude", fontsize=20)
g.savefig("B_data.jpeg")
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as ss
```

```python
R_data = pd.read_csv('Reduced-BLLac-Data-R.csv')
R_data_ATel = pd.read_csv('R-data-ATels.csv')
f = plt.figure()
f.set_figwidth(16)
f.set_figheight(9)
plt.errorbar(R_data.mjd, R_data.mag, yerr=R_data.mag_err, marker='o', ms=4, linestyle = "none", label = "LT data")
plt.errorbar(R_data_ATel.mjd, R_data_ATel.mag, yerr=R_data_ATel.mag_err, marker='s', ms=4, linestyle = "none", label = "ATel data")
plt.gca().invert_yaxis()
plt.legend()
plt.title("R-band", fontsize=30)
plt.xlabel("Modified Julian Date", fontsize=20)
plt.ylabel("Magnitude", fontsize=20)
f.savefig("R_data.jpeg")
plt.show()
```

```python
B_data = pd.read_csv('Reduced-BLLac-Data-B.csv')
g = plt.figure()
g.set_figwidth(16)
g.set_figheight(9)
plt.errorbar(B_data.mjd, B_data.mag, yerr=B_data.mag_err, marker='o', ms=4, linestyle="none")
plt.gca().invert_yaxis()
plt.title("B-band", fontsize=30)
plt.xlabel("Modified Julian Date", fontsize=20)
plt.ylabel("Magnitude", fontsize=20)
g.savefig("B_data.jpeg")
plt.show()
```

```python
Fermi_data = pd.read_csv('Fermidata.csv')
i = plt.figure()
i.set_figwidth(16)
i.set_figheight(9)
plt.errorbar(Fermi_data.MJD, Fermi_data.Flux, yerr=Fermi_data.Flux_err, marker='o', ms=4, linestyle='none')
plt.title("Fermi data", fontsize=30)
plt.xlabel("Modified Julian Date", fontsize=20)
plt.ylabel("Flux (photons/cm^2/s)", fontsize=20)
i.savefig("Fermi_data.jpeg")
plt.show()
```

```python
R_data2 = pd.read_csv('Reduced-BLLac-Data-R-2.csv')
Fermi_data2 = pd.read_csv('Fermidata_2.csv')
j = plt.figure()
j.set_figwidth(16)
j.set_figheight(9)
plt.errorbar(R_data2.mag, Fermi_data2.Flux, xerr=R_data2.mag_err, yerr=Fermi_data2.Flux_err, marker='o', ms=4, linestyle='none')
plt.gca().invert_xaxis()
plt.plot(np.unique(R_data2.mag), np.poly1d(np.polyfit(R_data2.mag, Fermi_data2.Flux, 1))(np.unique(R_data2.mag)))
P, C = ss.spearmanr(R_data2.mag, Fermi_data2.Flux)
plt.title("LT R band against Fermi", fontsize=30)
plt.xlabel("Magnitude", fontsize=20)
plt.ylabel("Flux (photons/cm^2/s)", fontsize=20)
j.savefig("R_band_Fermi.jpeg")
plt.show()
grad, intercept=np.polyfit(R_data2.mag, Fermi_data2.Flux, 1)
print("Gradient: " + str(grad))
print("Intercept: " + str(intercept))
print("C: " + str(C))
print("P: " + str(P))
```
